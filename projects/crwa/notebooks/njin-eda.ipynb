```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import os
import datetime
import sys
```

```python
data_dir = "../data/charles_river_samples_csv/"
df_dict = {file[:-4]:pd.read_csv(data_dir+file, encoding="latin-1") for file in os.listdir(data_dir)}
```

```python
for title, df in df_dict.items():
    display(title, df.head())
```

```python
with pd.option_context("display.max_columns", None):
    display(df_dict["Results"].head(30))
```

```python
results = df_dict["Results"]
```

```python
results["Time_Collected"] = pd.to_datetime(results["Time_Collected"])
```

```python
results["Time_Collected"] = results["Time_Collected"].dt.time
```

```python
dfs_to_join = [
    "Data_Type",
    "Media_Type",
    "Media_Subdivision",
    "Component",
    
    "Activity_Type",
    "Actual_Result_Type",
    "Reporting_Result_Type",
    "Relative_Depth",
]
```

```python
left_cols_to_join = [
    "Data_Type_ID",
    "Media_Type_ID",
    "Media_Subdivision_ID",
    "Component_ID",
    
    "Activity_Type_ID",
    "Actual_Result_Type_ID",
    "Reporting_Result_Type_ID",
    "Relative_Depth_ID",
]
```

```python
right_cols_to_join = [
    "Data_Type_ID",
    "Media_Type_ID",
    "Media_Subdivision_ID",
    "Component_ID",
    
    "Activity_Type_ID",
    "Result_Type_ID",
    "Result_Type_ID",
    "Relative_Depth_ID",
]
```

```python
for left_col, right_col, df_name in zip(left_cols_to_join, right_cols_to_join, dfs_to_join):
#     print(left_col, right_col, df_name)
    results = results.merge(df_dict[df_name], left_on=left_col, right_on=right_col)
```

```python
results.drop(columns=left_cols_to_join, inplace=True)
```
